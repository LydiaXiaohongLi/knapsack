```python
knapsack_raw=[]
with open(r'Data\knapsack1.txt','r') as f:
    knapsack_raw=f.readlines()
knapsack_item = [(int(x.split()[0]),int(x.split()[1])) for x in knapsack_raw[1:] ]
knapsack_capacity=int(knapsack_raw[0].split()[0])
```

```python
#value_array is a array with length of number of knapsack item 
#each cell of value_array stores a dictionary, dictionary with a key represents starting index, and with two attributes: length and value
#combinarion of the starting index and length represents the knapsack capacity of the current subproblem
#value represents the maximum knapsack value of the first i items (value_array index), given the knapsack capacity of the current subproblem
#knapsack algo: 
#wi weight of ith item, vi value of ith item, w knapsack capacity of current subproblem, W total knapsack capacity
#V(i,w) maximum value for first i items given current subpromble capacity w
#V(i,w)=max(V(i-1, w-wi)+vi, V(i-1,w))
value_array=[]
for i in range(len(knapsack_item)):
    #tmp_dict used to stores the dictionary to be populated for value_array[i]
    tmp_dict={}
    #initialize the value_array[0] for the first item, when knapsack capacity <= first item weight, then value is 0; else value is the first item value
    if i==0:        
        tmp_dict[0]={'length':knapsack_item[0][1]}
        tmp_dict[0]['value']=0
        tmp_dict[knapsack_item[0][1]]={'length':knapsack_capacity-knapsack_item[0][1]+1}
        tmp_dict[knapsack_item[0][1]]['value']=knapsack_item[0][0]
    else:
        #a is used to track the calculation of V(i-1,w)
        a=0
        a_value=value_array[i-1][a]['value']
        a_length=value_array[i-1][a]['length']
        #when w<=wi, V(i,w)=V(i-1,w)
        while a+a_length <= knapsack_item[i][1]:
            #index is used to track the V(i,w)
            index=a
            tmp_dict[index]={'length':a_length}
            tmp_dict[index]['value']=a_value
            a=a+a_length
            a_value=value_array[i-1][a]['value']
            a_length=value_array[i-1][a]['length']
        index=a
        tmp_dict[index]={'length':knapsack_item[i][1]-a}
        tmp_dict[index]['value']=a_value
        index_value=tmp_dict[index]['value']
        #b is used to track the calculation of V(i-1, w-wi)
        b=0
        b_left_length=value_array[i-1][b]['length']
        a_left_length=a+a_length-knapsack_item[i][1]
        while a <= knapsack_capacity:
            #V(i,w)=max(V(i-1, w-wi)+vi, V(i-1,w))
            tmp_value=max(value_array[i-1][b]['value']+knapsack_item[i][0],value_array[i-1][a]['value'])
            if b_left_length < a_left_length:
                #if the calculated knapsack value given is the same as last calcualted knapsack value
                #just need to extend the last knapsack value length, for the same knapsack value in the dictionary
                if tmp_value==index_value:
                    tmp_dict[index]['length']=tmp_dict[index]['length']+b_left_length
                #if the calculated knapsack value given is not the same as (bigger than) last calcualted knapsack value
                #create a new knapsack value entry in the dictionary                
                else:
                    index=index+tmp_dict[index]['length']
                    tmp_dict[index]={'length':b_left_length}
                    tmp_dict[index]['value']=tmp_value
                    index_value=tmp_value
                a_left_length=a_left_length-b_left_length
                b=b+value_array[i-1][b]['length']
                b_left_length=value_array[i-1][b]['length']
            else:
                if tmp_value==index_value:
                    tmp_dict[index]['length']=tmp_dict[index]['length']+a_left_length
                else:
                    index=index+tmp_dict[index]['length']
                    tmp_dict[index]={'length':a_left_length}
                    tmp_dict[index]['value']=tmp_value
                    index_value=tmp_value
                b_left_length=b_left_length-a_left_length
                a=a+value_array[i-1][a]['length']
                if a <= knapsack_capacity:
                    a_left_length=value_array[i-1][a]['length']               
    value_array.append(tmp_dict)
        
```
